### Sorting & Filtering

```python
df.sort_values('year', ascending=False, inplace=True)
```

```python
# the statement below simply assigns a T/F value to the record
df['year'] >= 2000

# the statement below filters the dataframe based on the T/F values
df[df['year'] >= 2000] 
```

```python
# Filter rows with year > 2010 and extract selected columns
df[df['year'] >= 2010][['year','make','model']]
```

```python
# Specify bounds
df[(df['year'] >= 2010) & (df['year'] <= 2012)].to_csv(r'datasets\filtered_cars.xlsx')
```

```python

l = ['BMW', 'Toyota']
for c in l:
    df[df['make']==c].to_csv(c+".csv")
```

```python
# extract a list of unqiue values in the make column
# delete nan values from make column in df  
# delete rows with nan values in make column
df = df.dropna(subset=['make'])
mylist = df['make'].unique()
print(mylist)
for m in mylist:
    df[df['make']==m].to_csv(m+".csv")
```

```python
import pandas as pd

# DataFrame 1: df1
df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Value_A': ['A', 'B', 'C', 'D']
})

# DataFrame 2: df2
df2 = pd.DataFrame({
    'ID': [3, 4, 5, 6],
    'Value_B': ['X', 'Y', 'Z', 'W']
})

print("left table: df1")
print(df1)

print("right table: df2")
print(df2)

# Inner Join: An inner join combines rows from two DataFrames based on a related column between them, eliminating rows that do not have matching values in both DataFrames.
print("inner join")
inner_joined_df = pd.merge(df1, df2, on='ID', how='inner')
print(inner_joined_df)

# Outer Join: An outer join returns all rows when there is a match in one of the DataFrames. Cells for which there are no matching data will contain NaN.
print("outer join")
outer_joined_df = pd.merge(df1, df2, on='ID', how='outer')
print(outer_joined_df)

# Left Join: A left join returns all rows from the left DataFrame, and the matched records from the right DataFrame. Non-matching records in the right DataFrame will be NaN.
print("left join")
left_joined_df = pd.merge(df1, df2, on='ID', how='left')
print(left_joined_df)

# Right Join: A right join returns all rows from the right DataFrame, and the matched records from the left DataFrame. Non-matching records in the left DataFrame will be NaN.
print("right join")
right_joined_df = pd.merge(df1, df2, on='ID', how='right')
print(right_joined_df)
```
