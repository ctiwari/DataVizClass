```python
import numpy as np
from scipy import stats
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(12345678)  # for reproducibility 
    
# Create two random distributions of data that may be related or not:
data_group1 = np.random.randn(50) + 8  # mean difference is 8, standard deviation is same as above 
data_group2 = np.random.randn(50) - 3   # mean difference is negative 3 and the standard deviations are same as above

# Performing two-sample t-test: parametric test (use scipy function for this)
t_stat, p_val1 = stats.ttest_ind(data_group1, data_group2)
print('T statistic and P value from SciPy T-Test Ind', t_stat, p_val1)  # prints the T statistic and two-sided pvalue 
    
# Performing nonparametric test (use statsmodels for this). Bootstrap resampling can be done using sm.statsdata class method called 'bootstrap'. It takes an instance of a rv_continuous or rv_discrete distribution as input, and return the bootstrapped statistic 
nonpara = sm.stats.DescrStatsW(data_group1).ttest_mean(data_group2)  # nonparametric test (use statsmodels library function for this)
print('Bootstrap T Stat & P Value', nonpara[0], nonpara[1])  
    
# Visualizing the data distributions:
plt.figure()  
sns.distplot(data_group1, label='Group 1')  # seaborn function to show histogram and fit line for given dataset (it will also add kde)
sns.distplot(data_group2, label='Group 2')  # we can use 'kde' argument here if it is False the plotting tool only shows a kernel density estimate of the underlying distribution as a curve. If True then shows both histogram and smooth curve behind it.
plt.legend()  
plt.show()  # show function to display all figures
```
